Regressao Linear


 Um modelo de regressão linear é construído para prever o 'Ano' com base na 'Contagem de passageiros'. Os dados são divididos em conjuntos de treinamento e teste usando train_test_split.

```python

a = ['Contagem de passageiros']
x = df.filter(items=a)
y = df_2['Ano']
```

```python
y
```

```python
print(f"Tamanho de X: {x.shape}")
print(f"Tamanho de y: {y.shape}")
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(x,
                                                    y,
                                                    test_size=0.25,
                                                    random_state=123)
```

```python
print(f"Tamanho de X_train: {X_train.shape}")
print(f"Tamanho de X_test: {X_test.shape}")
print(f"Tamanho de y_train: {y_train.shape}")
print(f"Tamanho de y_test: {y_test.shape}")
```

```python
from sklearn.linear_model import LinearRegression

clf = LinearRegression()
```

```python
clf.fit(X_train, y_train)
```

```python
clf.coef_.tolist()
```

Treinamento e avaliação do modelo: O modelo de regressão linear é treinado usando os dados de treinamento e avaliado usando os dados de teste. Métricas como Erro Quadrático Médio (MSE) e Erro Absoluto Médio (MAE) são calculadas para avaliar o desempenho do modelo.

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error

y_pred = clf.predict(X_test)

print(f"MSE de Teste: {mean_squared_error(y_test, y_pred):1.1f}")
print(f"MAE de Teste: {mean_absolute_error(y_test, y_pred):1.1f}")
```

Conceitos de MSE e MAE e como eles se aplicam ao cenário:

MSE (Erro Quadrático Médio):

Definição: O MSE calcula a média dos quadrados das diferenças entre os valores previstos pelo modelo e os valores reais.
Fórmula: MSE = (1/n) * Σ(y_real - y_previsto)^2
Interpretação: O MSE é uma medida da qualidade do modelo, onde valores menores indicam melhor desempenho. Como os erros são elevados ao quadrado, o MSE penaliza mais os erros grandes do que os erros pequenos.
Aplicação no cenário: No código, o MSE é usado para avaliar o desempenho do modelo de regressão linear na previsão do 'Ano' com base na 'Contagem de passageiros'. Um MSE menor indica que o modelo está fazendo previsões mais próximas dos valores reais do 'Ano'.
MAE (Erro Absoluto Médio):

Definição: O MAE calcula a média dos valores absolutos das diferenças entre os valores previstos pelo modelo e os valores reais.
Fórmula: MAE = (1/n) * Σ|y_real - y_previsto|
Interpretação: O MAE é outra medida da qualidade do modelo, onde valores menores indicam melhor desempenho. Ao contrário do MSE, o MAE trata todos os erros igualmente, sem penalizar mais os erros grandes.
Aplicação no cenário: No código, o MAE também é usado para avaliar o desempenho do modelo de regressão linear. Ele fornece uma medida da magnitude média dos erros de previsão do modelo em termos da unidade original da variável dependente ('Ano').
Comparação entre MSE e MAE:

Sensibilidade a outliers: O MSE é mais sensível a outliers (valores extremos) do que o MAE, pois os erros são elevados ao quadrado.
Interpretabilidade: O MAE é mais fácil de interpretar, pois é expresso na mesma unidade da variável dependente. O MSE, por outro lado, é expresso em unidades quadradas, o que pode ser menos intuitivo.
Escolha da métrica: A escolha entre MSE e MAE depende do objetivo da análise e da importância de penalizar erros grandes. Se a precisão geral do modelo for mais importante, o MSE pode ser uma boa escolha. Se a robustez a outliers for mais importante, o MAE pode ser preferível.
Conclusão:

No cenário do código, tanto o MSE quanto o MAE são usados para avaliar o desempenho do modelo de regressão linear na previsão do 'Ano'. O MSE penaliza mais os erros grandes, enquanto o MAE trata todos os erros igualmente. A escolha da métrica depende da importância de penalizar erros grandes e da interpretabilidade desejada. Em geral, é útil considerar ambas as métricas para obter uma visão completa do desempenho do modelo.
